## 🛠 Part 1: The Diagnostic Drill

```python
import numpy as np

# 1. The Identity Twist
diag_matrix = np.identity(5)
print("1. Identity Twist:\n", diag_matrix)
```

```python
# 2. Broadcasting Attack
vector = np.array([10, 20, 30, 40, 50])
broadcasted = vector + diag_matrix
print("\n2. Broadcasted Matrix:\n", broadcasted)
```

```python
# 3. The Threshold Filter
filtered = broadcasted.copy()
filtered[filtered > 25] = 99
print("\n3. Filtered Matrix (Threshold > 25):\n", filtered)
```

```python
# 4. Column Analysis
column_mean = np.mean(filtered[:, 2])
print("\n4. Mean of 3rd Column:", column_mean)
```

## 🌌 Part 2: The Gravity Grid (Simulation Base)
Building a coordinate system and calculating gravitational potential without triggering `RuntimeWarnings`.

```python
# 1. Create a coordinate grid
points = np.linspace(-10, 10, 5)
X, Y = np.meshgrid(points, points)

# 2. Calculate Distance R from Origin
R = np.sqrt(X ** 2 + Y ** 2)
print("Distance R from Origin:\n", R)
```

```python
# 3. Calculate Potential V = -1/R (Safely)
V = np.divide(-1, R, out=np.zeros_like(R), where=R!=0)
print("\nGravitational Potential V:\n", V)
```
